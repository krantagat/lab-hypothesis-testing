In this challenge, we will be working with pokemon data. The data can be found here:

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df.head(50)

df[df["Legendary"]==False]
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#code here

#Set hypothesis

#H0: Dragon average HP > Grass average HP
#H1: Dragon average HP < Grass average HP

#with alpha = 0.05

dragon_hp = df[(df["Type 1"]=="Dragon")|(df["Type 2"]=="Dragon")]["HP"]
grass_hp = df[(df["Type 1"]=="Grass")|(df["Type 2"]=="Grass")]["HP"]

dragon_hp
grass_hp
```

```python
alpha = 0.05
st.ttest_ind(dragon_hp,grass_hp, equal_var=False)
# accepet the null hypothesis
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
#code here
#Set hypothesis

#H0: Legendary HP, Attack, Defense, Sp.Def, Speed
# Grass average HP
#H1: Legendary average HP < Grass average HP

#with alpha = 0.05

# df[['Name', 'Age']]
legendary = df[(df["Legendary"]==True)][['HP', 'Attack','Defense','Sp. Atk','Sp. Def','Speed']]
# non_legendary = df[(df["Legendary"]==False)]
non_legendary = df[(df["Legendary"]==False)][['HP', 'Attack','Defense','Sp. Atk','Sp. Def','Speed']]


```

```python
display(non_legendary)
display(legendary)
```

```python
alpha = 0.05
st.ttest_ind(legendary,non_legendary, equal_var=True)
# reject the null hypothesis
```
